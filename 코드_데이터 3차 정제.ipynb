#### 국가대표어린이수영클럽 이름 변경

```python
national_swim_club_rows = df[df['시설명'] == '국가대표어린이수영클럽'].index
for row_index in national_swim_club_rows:
    df.at[row_index, '시설명'] = df.at[row_index, '시설명'] + ' ' + df.at[row_index, '자치구']
```
