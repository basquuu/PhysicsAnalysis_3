### 2. 서울시 수영장업 인허가 데이터

```python
allow = pd.read_csv('/content/2023_GYM/서울시 수영장업 인허가 정보.csv',encoding='cp949')
allow
```

#### 1. 필요 컬럼만 추출하기

```python
selected_columns = [ '사업장명','영업상태코드', '영업상태명', '상세영업상태명', '지번주소', '도로명주소', '데이터갱신일자', '데이터갱신구분']
```

```python
result = allow[selected_columns]
result
```

#### 2. 자치구 형태로 구분하기

```python
result['자치구'] = result['지번주소'].apply(lambda x: x.split(' ')[1] if isinstance(x, str) and '서울특별시' in x else None)
```

```python
result
```

```python
result['자치구'].value_counts()
```

#### 3. 상태영업상태명 확인

```python
result['상세영업상태명'].value_counts() #폐업 90개 정도로 생각보다 많음
```

- 폐업, 직권말소, 등록취소는 모두 삭제하기 , 휴업은 보류

```python
df = result[~result['상세영업상태명'].isin(['폐업', '직권말소', '등록취소'])]
df
```

#### 4. 컬럼 이름 통합하기

```python
# '상세영업상태명'을 '운영여부'로 칼럼 이름 변경
df.rename(columns={'상세영업상태명': '운영여부','사업장명':'시설명'}, inplace=True)
# '운영여부' 값 중 '영업중'인 경우를 1로, 그 외의 경우를 0으로 변환
df['운영여부'] = (df['운영여부'] == '영업중').astype(int)
df
```

- 불필요한 컬럼 삭제하고 index reset 하기

```python
df = df.drop(['영업상태명', '지번주소','데이터갱신구분','데이터갱신일자'], axis=1)
df = df.reset_index(drop=True)
df
```

```python
df['영업상태코드'].value_counts() #2 : 휴업인듯
```

#### 5. 데이터갱신일자 확인하기

- 코로나 19를 겪으며 수많은 수영장이 휴업 혹은 폐업하였다. 따라서, 코로나 19 이후 데이터 갱신이 이루어지지 않은 수영장의 경우 분석의 대상에서 제외하고 진행

```python
df['데이터 갱신일자'] = pd.to_datetime(df['데이터 갱신일자'], format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
# 2021년 이전인 행 추출
filtered_df = df[df['데이터 갱신일자'] > '2020-12-31']
filtered_df
```

### 4. 서울시 내 중, 고, 특수학교, 그 외, 각각 학교 데이터도 불러오기

```python
mid = pd.read_csv('/content/2023_GYM/2023년도_학생교육활동에 필요한 지원시설 현황(중)_서울특별시교육청.csv')
high = pd.read_csv('/content/2023_GYM/2023년도_학생교육활동에 필요한 지원시설 현황(고)_서울특별시교육청.csv')
special = pd.read_csv('/content/2023_GYM/2023년도_학생교육활동에 필요한 지원시설 현황(특)_서울특별시교육청.csv')
extra = pd.read_csv('/content/2023_GYM/2023년도_학생교육활동에 필요한 지원시설 현황(그)_서울특별시교육청.csv')
each = pd.read_csv('/content/2023_GYM/2023년도_학생교육활동에 필요한 지원시설 현황(각)_서울특별시교육청.csv')
```

```python
df_mid = mid[mid['수영장']== 1.0]
df_high = high[high['수영장']== 1.0]
df_special = special[special['수영장']== 1.0]
df_extra = extra[extra['수영장']==1.0]
df_each = each[each['수영장']==1.0]
#df_each, extra에는 수영장 없음
```

#### 1. 데이터 병합 진행

```python
merged_df = pd.concat([df_mid, df_high, df_special, df_extra, df_each], ignore_index=True)
```

```python
merged_df
```

#### 2. 자치구 컬럼 생성 , 필요 컬럼만 추출

```python
# '지역교육청'에서 '서울특별시' 이후만 추출해서 따로 저장
merged_df['지역 교육청'] = merged_df['지역교육청'].apply(lambda x: x.split('서울특별시')[1] if '서울특별시' in x else '')

# '지역'에서 '서울특별시' 이후 자치구만 따로 저장
merged_df['자치구'] = merged_df['지역'].apply(lambda x: x.split('서울특별시 ')[1] if '서울특별시' in x else '')
```

```python
df_tmp = merged_df[['지역 교육청','자치구', '학교명','설립구분','설립유형','수영장','체육관']]
df_tmp
```
